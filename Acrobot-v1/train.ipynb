```python
import gym
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import pickle
```

```python
def gather_data(env):

    sim_steps = 500
    num_sims = 5000
    num_successful_sims = 0
    trainX, trainY = [],[]

    # for num_sim in range(num_sims):
    while num_successful_sims < num_sims:
        observation = env.reset()
        training_sampleX, training_sampleY = [],[]

        for step in range(sim_steps):
            # env.render()
            action = np.random.randint(0, 3)
            one_hot_action = np.zeros(3)
            one_hot_action[action] = 1
            # print(one_hot_action)
            training_sampleX.append(observation)
            training_sampleY.append(one_hot_action)

            observation, reward, done, info = env.step(action) 

            if reward != -1:
                trainX += training_sampleX
                trainY += training_sampleY
                num_successful_sims += 1
                print(num_successful_sims, "/", num_sims, "IN STEP:", step)
                break
        

    trainX, trainY = np.array(trainX), np.array(trainY)

    return trainX, trainY
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(128, input_shape=(6,), activation="relu"))
    
    model.add(Dense(256, activation="relu"))
    model.add(Dropout(0.4))
    
    model.add(Dense(512, activation="relu"))
    model.add(Dropout(0.4))
    
    model.add(Dense(256, activation="relu"))
    model.add(Dropout(0.4))
    
    model.add(Dense(128, activation="relu"))
    model.add(Dense(3, activation="softmax"))
    
    model.compile(
        loss="categorical_crossentropy",
        optimizer="adam",
        metrics=["accuracy"])
    
    
    return model
```

```python
env = gym.make("Acrobot-v1")
trainingX, trainingY = gather_data(env)
```

```python
file_to_store = open("../models/Acrobot-v1/data/games2.pickle", "wb")
pickle.dump([trainingX, trainingY], file_to_store)
file_to_store.close()

```

```python
file_to_read = open("../models/Acrobot-v1/data/games2.pickle",'rb')
trainingX, trainingY = pickle.load(file_to_read)
file_to_read.close()
```

```python

model = create_model()
model.fit(trainingX, trainingY, epochs = 30, batch_size=100000)
model.save("Acrobot-v1-6th-try.h5")

```
